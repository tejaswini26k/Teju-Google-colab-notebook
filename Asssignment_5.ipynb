```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/cofee.csv')
df.plot()
plt.show()
```

```python
import matplotlib.pyplot as plt
#bar chart
country_bags=df.groupby('Country of Origin')['Number of Bags'].sum()
plt.bar(country_bags.index,country_bags.values)
plt.title("bar graph plot for each country origin has number of bags")
plt.xlabel("country_bags")
plt.xticks(rotation=45)
plt.ylabel("Number_of_bags")
```

```python
#stacked bar
Acidity=df.groupby('Country of Origin')['Balance'].sum()
Moisture_Percentage=df.groupby('Country of Origin')['Moisture Percentage'].sum()
plt.bar(Acidity.index,Acidity.values,label='Balance')
plt.bar(Moisture_Percentage.index,Moisture_Percentage.values,label='Moisture_Percentage')
plt.title("bar graph plot for each country origin has number of bags")
plt.xlabel("Balance")
plt.ylabel("Moisture Percentage")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
sweetness=df.groupby('Country of Origin')['Bag Weight'].sum()
moisture=df.groupby('Country of Origin')['Moisture Percentage'].sum()
plt.bar(sweetness.index,sweetness.values,label='Bag Weight')
plt.bar(moisture.index,moisture.values,bottom=sweetness.values,label='MoisturePercentage')
plt.xlabel('Country of Origin')
plt.ylabel('Value')
plt.title('Bag weight and Moisture Percentage for each Country of Origin')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
#scatter
plt.scatter(df['Flavor'],df['Aroma'])
plt.title('relation between flavor and sweetness')
plt.xlabel('Flavor')
plt.ylabel('Aroma')
plt.show()
```

```python
#line
plt.plot(df['Aftertaste'])
plt.title('change in aftertaste')
plt.xlabel('datapoint')
plt.ylabel('Aftertaste')
plt.show()
```

```python
#histogram
plt.hist(df['Moisture Percentage'])
plt.title('change in Moisture Percentage')
plt.xlabel('datapoint')
plt.ylabel('Aftertaste')
plt.show()
```
